```python
 import textstat
Readability_list_gunning_fog = []
```

```python
# #Readability_list_gunning_fog
import textstat
Readability_list_gunning_fog = []
for row in df4.iterrows():
        sentence = row[1]['Review_Text']
        if sentence == 0:
            readability_score = textstat.gunning_fog("0")
            Readability_list_gunning_fog.append(readability_score)
        else:
            readability_score = textstat.gunning_fog(sentence)

            Readability_list_gunning_fog.append(readability_score)
```

```python
# splitting the reviews 
df_split = df_HelpReduc.reset_index()
sentence_split = []
for i in df_split.iterrows():

    review = i[1]["Review_Text"]

    
    
    split_review = review.split(".")
    
    for j in split_review:
        start_list =[list(i[1].values)[0]]
        sentence_list = start_list
        sentence_list.append(j)
        sentence_split.append(sentence_list)
        

sentence_split
```

```python
#New dataframe to assign sentences
sentence_df = pd.DataFrame(sentence_split, columns=["ReviewID", "Sentence"])
```

```python
#find sentiment score for each document 
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
My_sentiment_analyser = SentimentIntensityAnalyzer()
```

```python
sentimental_text_list = []
for sentence in sentence_df["Sentence"]:

    text_score = My_sentiment_analyser.polarity_scores(sentence)['compound']
    sentimental_text_list.append(text_score) 
sentence_df["Sent"] = sentimental_text_list
```

```python
sentence_df
```

```python
sentence_df["PosSentLabel"] = sentence_df["Sent"].apply(lambda x: 1 if x >= 0.1 else 0)
sentence_df["NegSentLabel"] = sentence_df["Sent"].apply(lambda x: 1 if x <= -0.1 else 0)
sentence_df["Count"] = 1
```

```python
sentence_df.head(13)
```

```python
resultsDf = sentence_df.groupby('ReviewID').agg('sum')
```

```python
resultsDf
```

```python
resultsDf['PosRatio'] = resultsDf['PosSentLabel']/resultsDf['Count']
resultsDf['NegRatio'] = resultsDf['NegSentLabel']/resultsDf['Count']
resultsDf = resultsDf.reset_index()
```

```python
resultsDf
```

```python
sentence_label = []
for i in resultsDf.iterrows():
    posLab = 0
    negLab = 0
    if i[1]["PosRatio"] >= 0.1 and i[1]["NegRatio"] >= 0.1:
        posLab=1
  
    if posLab == 1:
        sentence_label.append(-1)
    else:
        sentence_label.append(1)
```

```python
resultsDf["TwoSided"] = sentence_label
```
